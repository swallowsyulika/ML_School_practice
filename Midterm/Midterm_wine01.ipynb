```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer 
```

```python
df = pd.read_csv(".\DATA\wine.csv")

# set 1
X_1 = df.iloc[:, 4].values.reshape(-1, 1)
y_1 = df.iloc[:, 10].values
# set 2
X_2 = df.iloc[:, 0].values.reshape(-1, 1)
y_2 = df.iloc[:, 9].values

# print(df.iloc[:, 4].isnull().values.any())    
# True

# deal with missing data
Set1Imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
Set1Imputer.fit(X_1[:, :])
X_1[:, :] = Set1Imputer.transform(X_1[:, :])

Set2Imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
Set2Imputer.fit(X_2[:, :])
X_2[:, :] = Set1Imputer.transform(X_2[:, :])

# split data
X_1_train, X_1_test, y_1_train, y_1_test = train_test_split(X_1, y_1, test_size=0.2, random_state=0)
X_2_train, X_2_test, y_2_train, y_2_test = train_test_split(X_2, y_2, test_size=0.2, random_state=0)
```

```python
# linear regression
Set1Regressor = LinearRegression()
Set1Regressor.fit(X_1_train, y_1_train)

Set2Regressor = LinearRegression()
Set2Regressor.fit(X_2_train, y_2_train)
```

```python
# print acc
S1acc = Set1Regressor.score(X_1_test, y_1_test)
print(f"Set1 acc: {S1acc}")

S2acc = Set2Regressor.score(X_2_test, y_2_test)
print(f"Set2 acc: {S2acc}")
```

```python
# plt set1 train
plt.scatter(X_1_train, y_1_train, c="blue")
plt.plot(X_1_train, Set1Regressor.predict(X_1_train), c="red")
plt.xlabel('Chlorides')
plt.ylabel('alcohol')
plt.show()
```

```python
# plt set1 test
plt.scatter(X_1_test, y_1_test, c="blue")
plt.plot(X_1_train, Set1Regressor.predict(X_1_train), c="red")
plt.xlabel('Chlorides')
plt.ylabel('alcohol')
plt.show()
```

```python
# plt set2 train
plt.scatter(X_2_train, y_2_train, c="blue")
plt.plot(X_2_train, Set2Regressor.predict(X_2_train), c="red")
plt.xlabel('Fixed acidity')
plt.ylabel('sulphates')
plt.show()
```

```python
# plt set2 test
plt.scatter(X_2_test, y_2_test, c="blue")
plt.plot(X_2_train, Set2Regressor.predict(X_2_train), c="red")
plt.xlabel('Fixed acidity')
plt.ylabel('sulphates')
plt.show()
```
