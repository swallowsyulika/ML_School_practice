```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv(".\DATA\PolynomialData.csv")

X = df.iloc[:,1].values
y = df.iloc[:,-1].values

X =  X.reshape(-1, 1)
print("Data split done!")
```

```python
X[:5]
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
# no fit ans
```

```python
plt.scatter(X, y, c="blue")
plt.plot(X, lin_reg.predict(X), c="red")
plt.xlabel('Data')
plt.ylabel('Predict')
plt.show()    
```

```python
# make poly nomial features
poly_reg = PolynomialFeatures(degree=3)    # degree=1 >> b + bX1  ==  linear
X_poly = poly_reg.fit_transform(X)

# linear gression
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)

plt.scatter(X, y, c="blue")
plt.plot(X, lin_reg2.predict(poly_reg.fit_transform(X)), c="red")
plt.xlabel('Data')
plt.ylabel('Predict')
plt.show()  
```

```python
X_poly  # degree = 3  ->  b + bx + bx^2 + bx^3 
```

```python
X_grid = np.arange(min(X), max(X), 0.1).reshape(-1, 1)    # smooth line
plt.scatter(X, y, c="blue")
plt.plot(X_grid, lin_reg2.predict(poly_reg.fit_transform(X_grid)), c="red")
plt.xlabel('Data')
plt.ylabel('Predict')
plt.show() 
```

```python
# predict with new data 6.5
new_x = np.array(6.5).reshape(-1, 1)
print(lin_reg.predict(new_x))
print(lin_reg2.predict(poly_reg.transform(new_x)))    # better
```
